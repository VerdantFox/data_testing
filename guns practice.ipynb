```python
def remove_non_county_parentheses(row):
    if 'county' in row['city']:
        return row['city']
    else:
        if '(' in row['city']:
            return row['city'].split(' (')[0]
        else:
            return row['city']
```

```python
guns['city'] = guns.apply(remove_non_county_parentheses, axis=1)
```

```python
def new_county_name(row):
    if 'county' in row['city']:
        _ = row['city'].split()[:-1]
        return ' '.join(_)
        
    else:
        if type(row['county_name']) == str:
            _ =  row['county_name'].replace('(', '')
            _ = _.replace(')', '')
            return _
        else:
            return None
        
```

```python
guns['county_name'] = guns.apply(new_county_name, axis=1)
```

```python
guns.head(50)
```

```python
def no_shots(row):
    if (row['n_killed'] == 0) and (row['n_injured'] == 0):
        return True
    else:
        return False
```

```python
guns['no_shots'] = guns.apply(no_shots, axis=1)
```

```python
guns.drop(guns[guns['no_shots'] == True].index, inplace=True)
guns.drop('no_shots', axis=1, inplace=True)
```
